```python
import random
```

```python
def lanzar(num_dados, iteraciones):
    caras, suma_resultados, resultados, dados, suma = [], [], [], [], 0
    for i in range(iteraciones):
        for j in range(num_dados):
            numero = random.randint(1,6)
            caras.append(numero)
            dados.append(numero)
            suma += numero
        resultados.append(dados)
        suma_resultados.append(suma)
        dados, suma = [], 0
    return resultados, suma_resultados, caras
```

```python
num_dados = int(input('¿Cuántos dados quieres lanzar?'))
iteraciones = int(input('¿Cuántos lanzamientos quieres lanzar?'))
resultados, suma_resultados, caras = lanzar(num_dados, iteraciones)
```

```python
max(suma_resultados)
```

```python
min(suma_resultados)
```

```python
import matplotlib.pyplot as plt
from statistics import mode
```

```python
plt.hist(suma_resultados, edgecolor="white")
plt.xlabel('Suma')
plt.ylabel('# de repeticiones')
plt.title('Suma de los resultados de los dados: ')
plt.show()
```

```python
print("La suma más frecuente fue: ", mode(suma_resultados))
```

```python
plt.hist(caras, edgecolor="black")
plt.xlabel('Número')
plt.ylabel('# de repeticiones')
plt.title('Número más frecuentes al lanzar los dados: ')
plt.show()
```

```python
print("El número más frecuente fue el: ", mode(caras))
```
